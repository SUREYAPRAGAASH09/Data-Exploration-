```python
import pandas as pd
import numpy as np

r_cols = ['user_id','movie_id','ratings']
filename_data = r"C:\Users\Sureya Pragaash\Downloads\Data for EDA\Movielens\ml-100k\u.data"
ratings = pd.read_csv(filename_data,sep = '\t',names = r_cols, usecols = range(3))

m_cols = ['movie_id','title']
filename_movies = r"C:\Users\Sureya Pragaash\Downloads\Data for EDA\Movielens\ml-100k\u.item"
movies = pd.read_csv(filename_movies,sep = '|', names = m_cols, usecols = range(2),  encoding='latin-1')

ratings = pd.merge(movies, ratings)

#ratings.head()

userRatings = ratings.pivot_table(index = ['user_id'], columns = ['title'], values = 'ratings' )

#corrMatrix = userRatings.corr()

corrMatrix = userRatings.corr(method = 'pearson', min_periods=1)

#INPUT
myRatings = userRatings.loc[2].dropna().head(10)

simCandidates = pd.Series()

for i in range(0, len(myRatings.index)):
    #print("Adding sims for "+ myRatings.index[i]+"...")
    # Retrive similar movies to this one that I rated
    sims = corrMatrix[myRatings.index[i]].dropna()
    # Now scale its similarity by how I rated this movie
    sims = sims.map(lambda x:x * myRatings[i])
    # Add the score to the list of similarity candidates
    simCandidates = simCandidates.append(sims)
    
#Glance at out resuts so far
#print ("sorting...")
simCandidates.sort_values(inplace = True, ascending = False)
#print(simCandidates.head(10))

simCandidates = simCandidates.groupby(simCandidates.index).sum()

#OUTPUT
simCandidates.sort_values(inplace = False, ascending = False).head(10)
```
